## Statistik Deskriptif dari Data Frame
Statistik deskriptif sebagai summary diperoleh dengan fungsi describe()
```
# dataframe describe
print([nama_dataframe].describe())
```

### Basic Statistik Deskriptif Data Frame 
memberikan informasi mengenai nilai rata-rata, standar deviasi dan interquartile range
- describe secara otomatis mengabaikan kolom kategori, cuma kolom numerik yang ditampilkan
- menambahkan parameter include="all" untuk mendapatkan summary dari numerik dan non numerik

```
# dataframe describe semuanya (numerik dan non numerik)
print([nama_dataframe].describe(include="all"))
```

### Statistik Deskriptif Data Frame pada Object
mendapatkan summary statistik untuk kolom non numerik dengan menambahkan parameter include="object"
```
# dataframe describe object
print([nama_dataframe].describe(include=["object"]))
```

### Statistik Deskriptif DataFrame mencari Tendensius
mencari nilai rata-rata, median, modus
```
# dataframe describe mencari rata-rata
print([nama_dataframe].loc[:, "nama_kolom"].mean())
# dataframe describe mencari nilai tengah
print([nama_dataframe].loc[:, "nama_kolom"].median())
# dataframe describe mencari nilai yang sering muncul
print([nama_dataframe].loc[:, "nama_kolom"].mode())
```

````
import pandas as pd

order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Quick summary  dari segi kuantitas, harga, freight value, dan weight
print(order_df.describe())

# Median median dari total pembelian konsumen per transaksi kolom price
print(order_df.loc[:, "price"].median())
````

## Mengenal dan Membuat Distribusi Data dengan Histogram
- cara untuk mengidentifikasi sebaran distribusi dari data
- grafik yang berisi ringkasan dari sebaran (dispersi atau variasi) suatu data
- tidak ada jarak antar batang/bar dari grafik karena titik data kelas bisa muncul di daerah cakupan grafik. 
  - Sedangkan ketinggian bar sesuai dengan frekuensi atau frekuensi relatif jumlah data di kelas. 
  - Semakin tinggi bar, semakin tinggi frekuensi data. Semakin rendah bar, semakin rendah frekuensi data

```
# syntax histogram
nama_dataframe[["nama_kolom"]].plot.hist(bins=jumlah_bins)
```
atribut
- bins = jumlah_bins, secara default jumlahnya 10
- by = kolom yang digroupby, value berupa nama column
- alpha = opacity dari plot, value dari 0.0 - 1.0 merupakan representasi ukuran opacity 
- figsize = ukuran dari plot, contoh: figsize=(10,12)

````
import pandas as pd
import matplotlib.pyplot as plt

order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# plot histogram kolom: price
order_df[["price"]].hist(figsize=(4, 5), bins=10, xlabelsize=8, ylabelsize=8)

# Untuk menampilkan histogram plot
plt.show()  
````

## Standar Deviasi dan Varians pada Pandas
standar deviasi merupakan suatu ukuran dispersi atau variasi yang 
- standar deviasi mempunyai satuan ukuran yang sama dengan satuan ukuran data asalnya
- varians  memiliki satuan kuadrat dari data asalnya (misalnya cm^2)

```
print([nama_dataframe].loc[:, "nama_kolom"].std())
print([nama_dataframe].loc[:, "nama_kolom"].var())
```

````
import pandas as pd

order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Standar variasi kolom product_weight_gram
print(order_df.loc[:, "product_weight_gram"].std())

# Varians kolom product_weight_gram
print(order_df.loc[:, "product_weight_gram"].var())
````

## Menemukan Outliers Menggunakan Pandas
Outliers, data observasi yang muncul dengan nilai-nilai yang paling jauh atau yang paling beda ama nilai sebagian besarnya

ditentukan dengan metric interquartile range (IQR)
```
# intequantile
Q1 = nama_df[["nama_kolom"]].quantile(0.25)
Q3 = nama_df[["nama_kolom"]].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```
rumus IQR = Q3 - Q1, memenuhi outlier kalau
  - data < Q1 - 1.5*IQR
  - data > Q3 + 1.5*IQR

memberikan output dengan beberapa nilai True atau False. 
- False yang berarti nilai-nilai ini valid 
- True menunjukkan adanya Outliers

```
# mencari outlier
print((nama_df < (Q1 - 1.5*IQR)) | (nama_df > (Q3 + 1.5*IQR)))
```

````
import pandas as pd
order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Hitung quartile 1
Q1 = order_df[["product_weight_gram"]].quantile(0.25)
# Hitung quartile 3
Q3 = order_df[["product_weight_gram"]].quantile(0.75)

# Hitung inter quartile range dan cetak ke console
IQR = Q3 - Q1
print(IQR)
````

## Groupby menggunakan Pandas
mencari summary dengan menggunakan aggregate dari kolom tertentu
```
# contoh syntax
nama_df["<kolom_dihitung>"].groupby([nama_df["<kolom_diagregat>"]]).mean()

# groupby mean
df["score"].groupby([df["name"]]).mean()
# groupby sum
df["score"].groupby([df["name"], df["exam"]]).sum()
```

````
import pandas as pd
order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Hitung rata rata dari price per payment_type
rata_rata = order_df["price"].groupby(order_df["payment_type"]).mean()
print(rata_rata)
````

## Sorting Menggunakan Pandas
metode mengurutkan data berdasarkan syarat kolom tertentu, digunakan untuk melihat nilai maksimum dan minimum dari dataset
```
# menggunakan fungsi sorting 
nama_df.sort_values(by="nama_kolom")
# sorting lebih satu kolom
# isi argument ascending sesuai urutan kolom
nama_df.sort_values(by=["nama_kolom1", "nama_kolom2"], ascending=[False, True])
```

````
import pandas as pd
order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Hitung harga maksimum pembelian customer
sort_harga = order_df.sort_values(by="price", ascending=False)
print(sort_harga)
````

````
import pandas as pd
import matplotlib.pyplot as plt

# import data
order_df = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/order.csv")

# Median price yang dibayar customer dari masing-masing metode pembayaran. 
median_price = order_df["price"].groupby(order_df["payment_type"]).median()
print("Median price yang dibayar customer dari masing-masing metode pembayaran : \n",median_price)

# Ubah freight_value menjadi shipping_cost dan cari shipping_cost 
# termahal dari data penjualan tersebut menggunakan sort.
order_df.rename(columns={"freight_value": "shipping_cost"}, inplace=True)
sort_value = order_df.sort_values(by="shipping_cost", ascending=0)
print("\nShipping Cost Termahal:\n",sort_value)

# Untuk product_category_name, berapa  rata-rata weight produk tersebut 
mean_value = order_df["product_weight_gram"].groupby(order_df["product_category_name"]).mean()
print("\nRata-rata weight produk berdasarkan kategori:\n",mean_value)
# dan standar deviasi mana yang terkecil dari weight tersebut
std_value = order_df["product_weight_gram"].groupby(order_df["product_category_name"]).std()
print("\nStandar Deviasi weight produk berdasarkan kategori:\n",std_value)

# Buat histogram quantity penjualan dari dataset tersebut untuk melihat persebaran quantity 
# penjualan tersebut dengan bins = 5 dan figsize= (4,5)
order_df[["quantity"]].hist(figsize=(4, 5), bins=5)
print("\nHistogram jumlah penjualan\n")
plt.show()
````
