# Pengantar
Visualisasi, cara merepresentasikan data melalui komponen visual, seperti posisi, komposisi, panjang, luas, dan warna. Serta cara berkomunikasi secara efektif kepada audiens melalui bahasa gambar

## Tujuan visualisasi data
 - Memecahkan masalah, bukan sekadar tampil cantik
   - memproses dan mengorganisasi fakta dan data, menemukan pola, hingga menghasilkan solusi
   - Desain visual membantu mengomunikasikan penyelesaian masalahnya
 - Fokus pada target dan audiens, bukan tools ataupun tekniknya. 
   - diperbolehkan menggunakan cara dan tools apa pun
   - tujuan menampilkan data tersebut dan siapakah audiensnya
 - Panduan dan alur, bukan asal menata gambar.
   - Urutan dan penempatan elemen visual harus dapat memandu audiens untuk memahami masalah hingga menuju jalan keluarnya
 - Konsep dan kerangka berpikir, bukan cuma kreativitas visual.
   - mentransfer konsep dan cara berpikir atas data dan proses analisisnya dengan cara yang mudah dipahami
   - meyakinkan audiens dengan konsep dan kerangka pikir yang dibangun dengan data sebagai bukti penguat
 - Kontekstual.
   - memiliki konteks yang sama dan terorganisasi dengan baik sehingga menambah kejelasan makna

## Eksploratif versus naratif
- Visual naratif
  - seringkali digunakan untuk presentasi, biasanya statis, dan berisi ringkasan informasi.
  - menjelaskan hasil akhir atau kesimpulan dari analis
  - penyajiannya cenderung menonjolkan keindahan visual
  - informasi tidak disampaikan secara detail
- Visual eksploratif 
  - cenderung menampilkan banyak data
  - butuh waktu lebih lama untuk mencerna kompleksitas data tersebut
  - lebih ditujukan untuk orang-orang yang memang memahami data

- visual naratif menyajikan dengan cerita dan pesan
  - tepat untuk disampaikan kepada pihak manager dan pihak umum
- visual exploratif menyajikan dengan analisis, hubungan sebab akibat perubahan data
  - tepat untuk disampaikan untuk orang di ranah IT

## Visualisasi Data 
visualisasi data tentang karyawan di perusahaan A

jumlah karyawan berdasarkan jenis kelamin dalam kelompok usia tertentu

### Visualisasi dengan Matlotlib
Komparasi jumlah karyawan laki-laki dan perempuan berdasarkan rentang usia

````
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use("seaborn")

tabel = (pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/usia_karyawan.csv").sort_values("Kelompok Usia", ascending=False).set_index("Kelompok Usia"))
tabel["Laki-laki"] = -tabel["Laki-laki"]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Perbandingan Jumlah Karyawan Laki-laki dan Perempuan\nBerdasarkan Kelompok Usia",
			 x=0., y=1.1, fontsize=24, fontweight='bold', ha='left')
tabel["Laki-laki"].plot(kind="barh", ax=axes[0], color="tab:blue", xlim=[-550, 0])
tabel["Perempuan"].plot(kind="barh", ax=axes[1], color="tab:orange", fontsize=14, xlim=[0, 550])
axes[0].set_ylabel("")
axes[0].set_xticklabels([""])
axes[0].set_yticklabels([""])
axes[0].legend(["Laki-laki"], fontsize=14, bbox_to_anchor=(0., 0., 1., .05))
axes[0].set_frame_on(False)
axes[1].set_ylabel("")
axes[1].set_xticklabels([""])
axes[1].legend(["Perempuan"], fontsize=14, bbox_to_anchor=(0., 0., 0.3, .05))
axes[1].set_frame_on(False)
for i, m, w in zip(range(tabel.shape[0]), list(tabel['Laki-laki']), list(tabel['Perempuan'])):
	axes[0].annotate(str(abs(m)), (m+20, i), xytext=(m+20, i), color="w", va="center", ha="center", fontsize=14)
	axes[1].annotate(str(abs(w)), (w-20, i), xytext=(w-20, i), color="w", va="center", ha="center", fontsize=14)
# untuk menyesuaikan parameter subplots untuk menentukan padding
plt.tight_layout()
plt.show()
````

mencari karyawan dengan tren produktivitas paling baik selama satu tahun lalu untuk mendapatkan reward sebagai best employee

Tujuan
- Memilih jenis grafik yang benar
- Menyederhanakan tampilan grafik
- Memilah dan menonjolkan informasi yang diperlukan

````
import pandas as pd
import matplotlib.pyplot as plt

tabel = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/produktivitas_karyawan.csv", index_col="Bulan")/1000

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
fig.suptitle("Best Employee 2019", x=0.028, y=1.05, fontsize=24, fontweight='bold', ha="left")
ax.set_title("Lima Karyawan dengan produktivitas tertinggi\n", fontsize=18, ha="right")

colors = ["tab:gray", "tab:gray", "tab:gray", "darkblue"]
line_widths = [1.5, 2, 1, 2.5, 3.]
lp = tabel.plot(kind="line",ax=ax, xlim=[0, 11], ylim=[10, 40], color=colors, 
				xticks=range(tabel.shape[0]), yticks=[10, 20, 30, 40])

add_lbl_pos = [0.3, 0.5, 1.2, -1.0, 0.5]
add_avg_pos = [-0.7, -0.5, 0.2, -2.0, -0.5]
for i, item in enumerate(lp.get_lines()):
	item.set_linewidth(line_widths[i])
	lbl = item.get_label()
	ax.annotate(lbl, (11, tabel.loc["Desember", lbl]), (11.2, tabel.loc["Desember", lbl] + add_lbl_pos[i]),
			   fontweight="bold", fontsize=12, color=colors[i], va="center",
			   arrowprops={"arrowstyle": "-", "color": colors[i]})
	ax.annotate("(avg: %.4f)" % tabel[lbl].mean(), (11, tabel.loc["Desember",lbl]),
			   (11.2, tabel.loc["Desember", lbl] + add_avg_pos[i]),
			   fontsize=11, color=colors[i], va="center")
highest_prod = tabel["Karyawan E"].max()
ax.annotate("Produktivitas\ntertinggi\n%.4f" % highest_prod, (8, highest_prod), (7.75, highest_prod+1.5),
		   ha="left", color=colors[-1], arrowprops={"arrowstyle": "-", "color": colors[-1]})
ax.set_xticklabels([item[:3] for item in tabel.index], fontsize=12)
ax.set_yticklabels([str(i) for i in [10, 20, 30, 40]], fontsize=12)
ax.set_xlabel("")
ax.set_ylabel("Jumlah produk (ribuan)\n", fontsize=14)
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
ax.spines["bottom"].set_color("none")
ax.get_legend().remove()
plt.show()
````

# Pengenalan Matplotlib dan Persiapan Dataset

## Pengenalan Dataset

````
import pandas as pd
dataset = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv')
print('Ukuran dataset: %d baris dan %d kolom\n' % dataset.shape)
print('Lima data teratas:')
print(dataset.head())
````

## Penambahan Kolom Order Month pada Dataset
membuat grafik GMV bulanan
- sumbu-x berupa bulan
- sumbu-y adalah total GMV



### menambahkan kolom order_month
membuat kolom baru order_month dari order_date

umumnya nilai x dari lambda mewakili nilai yang di assign

````
import datetime
# apply & lambda digunakan untuk membuat kolom baru, berdasarkan suatu kolom lain yang sudah ada
# dataset['order_date'] merupakan kolom yang sudah ada,  tiap nilai di dalamnya diproses menjadi month
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
print(dataset.head())
````

### menambahkan kolom GMV
GMV adalah perkalian setiap entri di kolom item_price dan kolom quantity

membuat kolom GMV dengan mengalikan kedua kolom yang disebutkan secara langsung

````
dataset['gmv'] = dataset['item_price']*dataset['quantity']
print('Ukuran dataset :%d baris dan %d kolom\n' % dataset.shape)
print('Lima data teratas:')
print(dataset.head())
````

# Plot Pertama dengan Matplotlib

## Membuat Data Agregat
agregasi pada jumlah gmv berdasarkan order_monthnya

````
monthly_amount = dataset.groupby('order_month')['gmv'].sum().reset_index()
print(monthly_amount)
````

## Membuat Line Chart Trend Pertumbuhan GMV
memanggil function plt.plot lalu definisikan nilai di sumbu-x dan sumbu-y
- order_month di sumbu-x (parameter pertama)
- kolom gmv di sumbu-y (parameter kedua). 
- plt.show()untuk menampilkan grafiknya

````
import matplotlib.pyplot as plt
plt.plot(monthly_amount['order_month'], monthly_amount['gmv'])
plt.show()
````

### Cara Alternatif: Fungsi .plot() pada pandas Dataframe
melakukan visualisasi dengan matplotlib tanpa membuat variabel baru

````
import matplotlib.pyplot as plt
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.show()
````

# Kustomisasi Grafik

## Anatomi Gambar Grafik
- Figure, keseluruhan visualisasi yang kita plot dalam 1 kali menjalankan code
- Axes, satuan plot yang dibuat pada saat memanggil .plot() / .scatter()
- komponen pada Axes (sebagai sumbu atau axis pada Y atau X)
  - Axis Label, nama sumbu yang ditampilin
  - Tick, titik garis penanda yang sejajar ama sumbu
  - Tick Label, tulisan tiap tick yang menyatakan nilainya
- grafik yang dipanggil biasa berupa line atau marker
  - .plot(), memanggil grafik line / line plot
  - .scatter(), memanggil grafik scatter/ marker
  - .bar(), memanggil grafik bar/ histogram
- Grid, mempermudah melihat garis sejajar
- Text, memberikan informasi berbentuk text

## Mengubah Figure Size
```
# mengubah gambar
plt.figure(figsize = (<panjang>, <lebar>))
```

````
import matplotlib.pyplot as plt
# atur figure size
plt.figure(figsize=(15,5))
dataset.groupby(['order_month'])['gmv'].sum().plot()
plt.show()
````

## Menambahkan Title and Axis Labels
- title, judul  
- axis labels, label di sumbu-x & sumbu-y 

````
import matplotlib.pyplot as plt
 
# atur ukuran
plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['gmv'].sum().plot()

# menambah plt.title, plt.xlabel, dan plt.ylabel sebelum plt.show()
plt.title('Monthly GMV Year 2019')
plt.xlabel('Order Month')
plt.ylabel('Total GMV')
plt.show()
````

## Kustomisasi Title and Axis Labels
parameter yang bisa ditambah
- loc,  menentukan posisi title (defaultnya title ada di tengah) 
  - ‘left’ untuk membuat rata kiri
  - ‘right’ untuk rata kanan
  - ‘center’ untuk meletakkannya di tengah
- pad, menambahkan jarak antara judul ke grafik (dalam satuan px)
- fontsize, mengganti ukuran font/huruf (dalam satuan px)
- color, mengganti warna huruf judul
  - warna dasar dengan kata seperti ‘blue’, ‘red’, ‘orange’, dsb. 
  - hex string, misalnya '#42DDF5' untuk warna biru muda

````
import matplotlib.pyplot as plt

# atur ukuran
plt.figure(figsize=(15, 5))

dataset.groupby(['order_month'])['gmv'].sum().plot()

# menambah judul , trus di kostumisasi
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')

# menambah label
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount', fontsize=15)
plt.show()
````

## Kustomisasi Line dan Point
parameter yang bisa dikustomisasi
- color, mengubah warnanya (sama seperti di title)
- linewidth, mengubah ketebalan garisnya (dalam satuan px)
- linestyle, mengubah jenis dari garis. 
  - '-' atau 'solid' untuk garis tak terputus (seperti pada default)
  - '--' atau 'dashed' untuk garis putus-putus
  - ':' atau 'dotted' untuk garis berupa titik-titik
  - '-.' atau 'dashdot' untuk garis dan titik bergantian  
- marker, mengubah tipe points/titik data di chart
  - '.' untuk bulatan kecil/titik
  - 'o' untuk bulatan agak besar, 's' untuk persegi
  - 'D' untuk diamond/wajik
  - bentuk-bentuk lain seperti '+', 'x', '|', '*'


````
import matplotlib.pyplot as plt

# atur ukuran gambar
plt.figure(figsize=(15, 5))

# plot dan kostumisasi pada garis grafik dan titiknya
dataset.groupby(['order_month'])['gmv'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)

# judul dan kostumisasi
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')

# label pada sumbu
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount', fontsize=15)

# tampilin gambar
plt.show()
````

## Kustomisasi Grid
agar nilai di masing-masing titik bisa mudah dipetakan ke sumbu-y

dengan menambahkan garis

````
import matplotlib.pyplot as plt

# atur ukuran gambar
plt.figure(figsize=(15, 5))

# kustomisasi grafik pada garis dan titik
dataset.groupby(['order_month'])['gmv'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)

# kostumisasi judul
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')

# nambahin label
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount', fontsize=15)

# nambahin garis
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.show()
````

## Kustomisasi Axis Ticks
mengakses nilai di sumbu x dan y melalui function plt.xticks() dan plt.yticks()

buat dalam bentuk miliaran kalau mudah dipahami

````
import matplotlib.pyplot as plt

# atur ukuran
plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount (in Billions)', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)

# kostumisasi untuk mengakses nilai pada sumbu x dan sumbu y
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.show()
````

## Menentukan Batas Minimum dan Maksimum Axis Ticks
- menentukan nilai pada masing - masing sumbu
  - dengan fungsi ylim(ymin=0)
- menentukan batas maksimum
  - dengan parameter ymax
- mengatur batas minimum dan maksimum sumbu-x 
  - dengan function plt.xlim

```
# batas sumbu x
plt.xlim(xmin=<batas_awal>, xmax=<batas_akhir>)
# batas sumbu y
plt.ylim(ymin=<batas_awal>, ymax=<batas_akhir>)
```

````
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount (in Billions)', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
# menambahkan batas minumum dengan ymin
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.show()
````

## Menambahkan Informasi Pada Plot
menambahkan informasi pada grafik agar pembaca mendapatkan insight 

parameter yang diset pada plt.text
- koordinat, x dan y. 
  - saat set transform=fig.transFigure, maka koordinatnya berkisar 0 sampai 1
  - untuk x dari kanan ke kiri
  - untuk y dari bawah ke atas
- Jika parameter transform tidak diisi, maka koordinatnya dalam satuan inch

```
plt.text(<kordinat_x>, <kordinat_y>, 
        '<tulisan>', transform=fig.transFigure, color='red')
```

mulai bulan Oktober 2019, GMV kita naik drastis setelah sebelumnya stagnan di angka 200-300 milyar per bulan

````
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['gmv'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)
plt.title('Monthly GMV Year 2019', loc='center', pad=40, fontsize=20, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount (in Billions)', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
# menambahkan informasi pada plot dengan plt.text
plt.text(0.45, 0.72, 'The GMV increased significantly on October 2019', transform=fig.transFigure, color='red')
plt.show()
````

- facecolor, Memberikan warna bagian depan figure, di luar area plot 
- edgecolor, Memberikan warna pinggiran gambar
- transparent, Jika nilainya True, maka gambarnya jadi transparan (jika filenya png)

- Sumbu-y adalah jumlah unique customers di tiap tanggal
- Title dan axis label harus ada, tulisan dan style-nya silakan disesuaikan sendiri

````
# Import library yang dibutuhkan
import datetime
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset retail_raw_reduced.csv
dataset = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv')
# Buat kolom order_month
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
# Buat kolom gmv
dataset['gmv'] = dataset['item_price'] * dataset['quantity']

# atur ukuran gambar
plt.figure(figsize=(10, 5))
# variable yang ditampilkan
dataset[dataset['order_month']=='2019-12'].groupby(['order_date'])['customer_id'].nunique().plot(color='red', marker='.', linewidth=2)
# kustomisasi judul
plt.title('Daily Number of Customers - December 2019', loc='left', pad=30, fontsize=20, color='orange')
# label pada sumbu x dan y
plt.xlabel('Order Date', fontsize=15, color='blue')
plt.ylabel('Number of Customers', fontsize=15, color='blue')
# garis pada grafik
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
# batas minimum pada axis ticks
plt.ylim(ymin=0)
# tampilkan grafik
plt.show()
````

# Jenis-Jenis Visualisasi Data
- Perbandingan/Komparasi
  - membandingkan suatu sekumpulan nilai dengan nilai lainnya, antara yang lebih kecil dan lebih besar
    - line-chart, membandingan data yang terkait dengan waktu
    - bar-chart, membandingkan kategorikal  
- Distribusi
  - melihat distribusi/persebaran data di suatu variabel
  - menggunakan histogram
- Komposisi
  -  melihat komposisi dari suatu variabel, jika di-breakdown terhadap suatu dimensi data
  - Visualisasi yang biasa digunakan
    - stacked bar-chart untuk data kategorikal
    - stacked line-chart untuk data terkait waktu
- Keterhubungan/Relasi
  - melihat keterhubungan antara suatu variabel
  - misalnya, jika variabel A semakin tinggi, 
    - apakah variabel B juga semakin tinggi? 
    - atau justru lebih rendah ? 
    - atau tidak ada keterhubungannya ?
  - Visualisasi yang biasa digunakan adalah scatter-plot

## Multi-line Chart
menampilkan banyak garis dalam grafik berdasarkan kategrorinya

breakdown datanya berdasarkan dimensi berdasarkan brand atau merk

hasil GMV dipantau perbulan

````
# Import library
import datetime
import pandas as pd
import matplotlib.pyplot as plt
# Baca dataset
dataset = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv')
# Buat kolom baru yang bertipe datetime dalam format '%Y-%m'
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
# Buat Kolom GMV
dataset['gmv'] = dataset['item_price']*dataset['quantity']

# Buat Multi-Line Chart
# group by berdasarkan order perbulan dan brand, trus gmv di sum unstack
dataset.groupby(['order_month', 'brand'])['gmv'].sum().unstack().plot()
plt.title('Monthly GMV Year 2019 - Breakdown by Brand', loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Order Month', fontsize = 15)
plt.ylabel('Total Amount (in Billions)', fontsize = 15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
# atur ukuran gambar dalam satuan inci
plt.gcf().set_size_inches(10, 5)
# 
plt.tight_layout()
plt.show()
````

### Kustomisasi Legend
Atur komponen legend menggunakan function plt.legend()
```
plt.legend(loc='<>', bbox_to_anchor=(<>, <>), shadow=<>, ncol=<>)
```

parameter yang bisa ditambahkan
- loc: untuk menentukan posisi legend
  - 'upper left', legend diletakkan di pojok dari atas kiri axes
  - 'upper right', legend diletakkan di pojok dari atas kanan axes
  - 'lower left', legend diletakkan di pojok dari bawah kiri axes
  - 'lower right', legend diletakkan di pojok dari atas kiri axes 
  - 'upper center', legend diletakkan di tepi atas tengah axes
  -'lower center', legend diletakkan di tepi bawah tengah axes
  - 'center left', legend diletakkan di tepi tengah kiri axes
  - 'center right', legend diletakkan di tepi tengah kanan axes 
  - 'center': legend diletakkan di tengah-tengah axes
  - 'best': matplotlib memilih lokasi legend terbaik agar tidak overlap dengan isi grafik
- bbox_to_anchor: biasanya digunakan untuk adjust lokasi dari legend. 
  - Bisa berisi 2 angka yang menunjukkan koordinat x dan y (misal (1.6,0.5) berarti geser 1.6 ke kanan dan 0.5 ke atas). 
  - Bisa juga berisi 4 angka, angka ketiga dan keempat menyatakan width (lebar) dan height (tinggi) dari legend.
- shadow: jika diisi True, maka kotak legend akan memiliki bayangan.
- ncol: jumlah kolom dari isi legend, defaultnya adalah 1
- fontsize: ukuran huruf pada legend
- title: memberikan judul pada legend
- title_fontsize: ukuran huruf pada judul legend

````
import matplotlib.pyplot as plt
dataset.groupby(['order_month','brand'])['gmv'].sum().unstack().plot()
plt.title('Monthly GMV Year 2019 - Breakdown by Brand',loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Order Month', fontsize = 15)
plt.ylabel('Total Amount (in Billions)',fontsize = 15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
# kostumisasi legend
plt.legend(loc='right', bbox_to_anchor=(1.6, 0.5), shadow=True, ncol=2)
plt.gcf().set_size_inches(12, 5)
plt.tight_layout()
plt.show()
````
